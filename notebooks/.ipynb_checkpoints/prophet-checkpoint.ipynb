```python
def plot_grid(df, n_cols, figsize):
    n_rows = int(np.ceil(len(df.columns)/n_cols))
    df.plot(subplots=True, layout=(n_rows, n_cols), figsize=figsize)
```

```python
from fbprophet import Prophet
from sklearn.model_selection import train_test_split
import statsmodels.tsa.api as tsa
pd.plotting.register_matplotlib_converters()
```

```python
y = forecastable[16070]
y.index.name = "ds"
y.name = "y"
y.plot();
```

```python
train, test = y[:"2018-12-31"], y["2019-01-01":]
```

# Prophet

```python
m = Prophet(seasonality_mode="multiplicative", 
            yearly_seasonality=True, 
            weekly_seasonality=False, 
            daily_seasonality=False, 
            interval_width=0.8, 
            mcmc_samples=1000)
m.fit(train.reset_index());
future = m.make_future_dataframe(365, freq="D")
fcst = m.predict(future)
fcst = fcst.set_index("ds")
```

```python
ax = fcst.yhat.plot()
plt.fill_between(fcst.index, y1=fcst.yhat_lower, y2=fcst.yhat_upper, alpha=0.3)
train.plot(style=".")
test.plot(style=".")
```

```python
fcst.yhat.resample("M").sum().plot()
train.resample("M").sum().plot()
test.resample("M").sum().plot()
```
